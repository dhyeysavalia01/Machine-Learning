### Array Fundamentals

It is mutable

```python
import numpy as np
```

```python
# initializing an array.

arr = np.array([1,2,3,4,5])
arr
```

```python
arr[0]
```

```python
# updating array elements
arr[0] = 100
print(arr)
```

```python
# replacing given elements with another particular element
arr[:3] = 22
arr
```

```python
type(arr)
```

### Create a basic array

```python
# creating array of Zeros
zeros_array = np.zeros(7)
zeros_array
```

```python
# array of one
ones_array = np.ones(6)
ones_array
```

```python
# creating empty array with 3 elements
empty_array = np.empty(3)
print(empty_array)

print(empty_array[0])
```

```python
# array with range of elements
# (starting, ending, step)
range_Array = np.arange(1,6,2)
range_Array
```

```python
# create an array with values that are spaced linearly in a specified interval
linearly_spaced_Array = np.linspace(2,10,5)
linearly_spaced_Array
# NOTE : here the last element is included in output array.
```

```python
# specifying datatype during array creation
custom_Array = np.arange(4,14,1, dtype = np.float64)
custom_Array
```

### Adding, removing, and sorting elements

```python
import random as rd

random_Array1 = [rd.randint(10,99)for _ in range(7)]
random_Array1
```

```python
# sorting an array
np.sort(random_Array1)
```

```python
random_Array2 = [rd.randint(10,99)for _ in range(5)]
random_Array2
```

```python
np.concatenate((random_Array1,random_Array2))
```

```python
new_arr1 = np.arange(10).reshape(2,5)
new_arr1
```

```python
new_arr2 = np.arange(10,20).reshape(2,5)
new_arr2
```

```python
concatenated_Array = np.concatenate((new_arr1,new_arr2))
concatenated_Array
```

```python
concatenated_Array.shape
```

```python
concatenated_Array.size
```

```python
concatenated_Array.ndim
```

### Reshaping an Array

```python
x = np.array([2,4,6,8,10,12])
print(x)
print(x.shape)
print(x.size)
print(x.ndim)
```

```python
y = x.reshape(3,2)
print(y)
print('\n')
print(y.shape)
print(y.size)
print(y.ndim)
```
