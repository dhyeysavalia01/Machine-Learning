## Implement dot product manually vs with NumPy

### Manual implementation

```python
# Manual implementation: dot product
# vectors


a = [1, 2, 3]
b = [4, 5, 6]

dot = 0

for i in range (len(a)):
    dot += a[i]*b[i]

print(dot)

```

```python
# Manual implementation: dot product
# matrices


A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

result = [[0,0],[0,0]]

for i in range (len(A)):
    for j in range (len(B[0])):
        for k in range (len(B)):
            result[i][j] += A[i][k] * B[k][j]

print(result)
```

### NumPy implementation

```python
# dot product : vectors
import numpy as np

a = [1, 2, 3]
b = [4, 5, 6]

print(np.dot(a,b))
```

```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

print(np.dot(A,B))
```
