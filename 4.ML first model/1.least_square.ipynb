# <span style='color:yellow'>Linear regression - Least Square Method</span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample Dataset
x = np.array([600,800,1000,1200,1400]) # house size in sq.ft
y = np.array([32,42,52,62,72]) # price in lakhs
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

print(x_mean)
print(y_mean)
```

```python
numerator = np.sum((x_mean - x)*(y_mean - y))
denominator = np.sum((x_mean - x)**2)

m = numerator/denominator
print(m)
```

```python
intercept_b = y_mean - (m*x_mean)
print(intercept_b)
```

```python
print(f"Line of regression: y = {m}x + {intercept_b}")
```

```python
# prediction on training data itself
pred_on_training_data = (m*x)+intercept_b
print(pred_on_training_data)
```

```python
# prediction in input 1600
new_x = 1600
new_pred = m*new_x + intercept_b
new_pred
```

```python
# the plot
plt.scatter(x,y,color='blue',label='actual data')
plt.plot(x,pred_on_training_data,color='red',label='fitted line')
plt.xlabel('house size in sq.ft')
plt.ylabel('price in lakhs')
plt.title('Linear Regression (Least Squares)')
plt.legend()
plt.grid(axis='both',alpha = 0.7)
plt.show()
```
