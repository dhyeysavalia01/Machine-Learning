# <span style='color:yellow'> Gradient Descent </span>
<span style='color:yellow'> single parameter : intercept </span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample dataset
x = np.array([0.5,2.3,2.9]) # weight
y = np.array([1.4,1.9,3.2]) # height
# points a,b,c
```

```python
# plt.figure(figsize=(8,6))
plt.scatter(x,y,color='blue',label='actual data',s=70)
plt.xlabel('weigth')
plt.ylabel('height')
plt.grid(axis='both', ls = 'dashed')
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)


# slope, from least square estimate
numerator = np.sum((x_mean - x)*(y_mean - y))
denominator = np.sum((x_mean - x)**2)

slope = numerator/denominator
print(f"slope: {slope:.3f}")
```

```python
# assuming the parameter
intercept = -5

# the hyper-parameters
steps = 1000
learning_rate = 0.01
n = len(x) # 3
```

```python
for i in range(steps):
    y_pred = intercept + slope*x
    residuals = y - y_pred      # actual - predicted
    
     # SSR loss function : np.mean((y - (intercept + slope*x))**2)
     # derivative of SSR loss function, giving the slope
    gradient = (-2/n)*np.sum(residuals)
    step_size = learning_rate * gradient
    intercept = intercept - step_size

    
    print(f"step:{i}, intercept:{intercept}, step size:{step_size}")
    
```

```python
# After the loop
y_pred = intercept + slope * x  # <-- recalculate using final intercept

# Plot
plt.scatter(x, y, color='blue', label='Actual Data', s=70)
plt.plot(x, y_pred, color='red', label='Fitted Line')
plt.xlabel('weight')
plt.ylabel('height')
plt.legend()
plt.grid(axis='both', ls='dashed')
plt.title('Final Fitted Line')
plt.show()
```
