# Simple linear regression : Gradient Descent.
### <span style='color:yellow'>Assuming an arbitary slope and intercept both.</span>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv')
df.head()
```

```python
x = df['house_size'].values
y = df['price_usd'].values

# normalizing the dataset, else gradient will explode.
x = (x - np.mean(x))/np.std(x)
y = (y - np.mean(y))/np.std(y)
```

```python
plt.scatter(x,y,color = 'salmon', label = 'actual data',s = 100, edgecolor = 'red')
plt.xlabel('house size')
plt.ylabel('price usd')
plt.grid(True)
plt.show()
```

```python
# Defining hyperparameters
m = 0
b = 0

steps = 1000
lr = 0.01
n = len(x)

# tracking loss over each step
residuals_list = []
```

```python
for i in range(steps):
    y_pred = m*x + b
    residuals = y - y_pred  # actual - predicted
    
    loss_fn = np.mean((residuals)**2)
    residuals_list.append(loss_fn)

    # gradients of each
    gradient_residuals = (-2/n)*np.sum(residuals)
    gradient_intercept = (-2/n)*np.sum(x * residuals) 

    step_size_res = lr * gradient_residuals
    step_size_int = lr * gradient_intercept

    b = b - step_size_res
    m = m - step_size_int
    print(f"step {i}, slope = {m}, intercept = {b}, loss = {loss_fn}")
        
```

```python
# Final Results
print("\n Final Slope (m):", round(m, 4))
print(" Final Intercept (b):", round(b, 4))
```
