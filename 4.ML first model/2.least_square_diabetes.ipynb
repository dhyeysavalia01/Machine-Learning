# <span style='color:yellow'>Linear regression - Least Square Method</span>

<span style="color:yellow">Diabetes dataset</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('diabetes.csv')
df.head()
```

```python
# selsecting the input features and output variable
x = df['BMI'].values
y = df['Glucose'].values
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

print(x_mean)
print(y_mean)
```

```python
# finding the slope
numerator = np.sum((x_mean - x)*(y_mean - y))
denominator = np.sum((x_mean - x)**2)
m = numerator/denominator
print(m)
```

```python
# finding the b intercept
b = y_mean - m*x_mean
print(b)
```

```python
print(f"line of regression: y = {m:.3f}x + {b:.3f}")
```

```python
# prediction on training data
y_pred = m*x + b
print(y_pred)
```

```python
# prediction on glucose = 45
new_x = 45
new_pred = m*new_x + b
print(new_pred)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y,y_pred)
mae
```

```python
from sklearn.metrics import r2_score
print(f"R² Score: {r2_score(y, y_pred):.3f}")
```

```python
plt.scatter(x,y,color = 'blue', label = 'Actual values')
plt.plot(x,y_pred,color='red',label = 'fitted line')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.legend()
plt.title('Linear Regression: BMI vs Glucose')
plt.show()
```
