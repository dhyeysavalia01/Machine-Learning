# MULTIPLE LINEAR REGRESSION : Least Square method

## Diabetes dataset
## 4 inputs, 1 output

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
# inputs : Glucose, BloodPressure, SkinThickness, BMI
# output : DiabetesPedigreeFunction
x1 = df['Glucose'].values
x2 = df['BloodPressure'].values
x3 = df['SkinThickness'].values
x4 = df['BMI'].values
y = df['DiabetesPedigreeFunction'].values
```

```python
# standardizing the data
x1 = (x1 - np.mean(x1)) / np.std(x1)
x2 = (x2 - np.mean(x2)) / np.std(x2)
x3 = (x3 - np.mean(x3)) / np.std(x3)
x4 = (x4 - np.mean(x4)) / np.std(x4)
y = (y - np.mean(y)) / np.std(y)
```

```python
x = np.column_stack((np.ones(len(x1)),x1,x2,x3,x4))
x
```

```python
p = x.T @ x
p_inv = np.linalg.inv(p)
q = x.T @ y

b = np.dot(p_inv,q)
b
```

```python
b0,b1,b2,b3,b4 = b
```

```python
# prediction on training data
y_pred = x @ b
y_pred
```

```python
print(y)
```

```python
# the loss
mse = np.mean((y-y_pred) ** 2)
mse
```

```python
# the plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111,projection = '3d')

# plotting actual data and predicted data
ax.scatter(x1,x2,y,color = 'blue',s = 70, label = 'actual data')
ax.scatter(x1,x2,y_pred,color = 'red',s = 70, label = 'prediction training data')

# the grids
x1_grid, x2_grid = np.meshgrid(np.linspace(min(x1),max(x1),10), np.linspace(min(x2),max(x2),10))
x3_fixed = np.mean(x3)
x4_fixed = np.mean(x4)
y_grid = b0 + b1*x1_grid + b2*x2_grid + b3*x3_fixed + b4*x4_fixed

# the plot
ax.plot_surface(x1_grid,x2_grid,y_grid, color = 'yellow',alpha = 0.5, label = 'Regression Plane')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('MLR : 4:1')
plt.legend()
plt.show()
```
