```python
# Mean, Variance, Std.deviation

def mean(values):
    return sum(values)/len(values)


def variance(values):
    total = 0
    n = len(values)

    for i in values:
        total = total + ((i-mean(values))**2)

    return (total/n)


def std_dev(values):
    return (variance(values)**0.5)


values = [4, 8, 6, 5, 3, 7, 10, 6]

print(mean(values))
print(variance(values))
print(std_dev(values))


```

```python
# Meab Squared error

# Actual (true) values
y_true = [3, -0.5, 2, 7, 4.2]

# Predicted values
y_pred = [2.5, 0.0, 2, 8, 5.3]


def mse(y_true, y_pred):
    n = len(y_true)
    sumation = 0

    for i,j in zip(y_true,y_pred): # applied zip function.
        sumation = sumation + ((i - j)**2)

    return (sumation/n)

print(mse(y_true,y_pred))

```
