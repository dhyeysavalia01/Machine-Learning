# <span style='color:orange'>Functions</span>

A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.

A function can return data as a result.

```python
def say_hello():
    print("hello everyone")

say_hello()
```

```python
name = input("Enter name :")

def f2(name):
    print("hello, "+name)

f2(name)
```

### <span style='color:orange'>Arbitrary Arguments, *args</span>

<span style='color:yellow'>If the number of arguments is unknown, add a * before the parameter</span>

```python
def f3(*details):
    print("Name:\t",details[0])
    print("PRN:\t",details[1])
    print("Branch:\t",details[2])
    
f3('Dhyey',1032220948,'CSE')
```

### <span style='color:orange'>Keyword Arguments</span>

<span style='color:yellow'>You can also send arguments with the key = value syntax.</span>

<span style='color:yellow'>This way the order of the arguments does not matter.</span>

```python
def f4(name,prn,branch):
    print("Name:\t",name)
    print("PRN:\t",prn)
    print("Branch:\t",branch)

f4(prn=1032220948,branch="CSE",name='Dhyey')
```

### <span style='color:orange'>Arbitrary Keyword Arguments, **kwargs</span>

<span style='color:yellow'>If you do not know how many keyword arguments that will be passed into your function, add two asterisk: ** before the parameter name in the function definition.</span>



```python
def f5(**details):
    print("Name:\t",details['NAME'])
    print("PRN:\t",details['PRN'])
    print("Branch:\t",details['BRANCH'])

f5(NAME='Dhyey',BRANCH='CSE',PRN=1032220948)
```

### <span style='color:orange'>Default Parameter Value</span>

<span style='color:yellow'>If we call the function without argument, it uses the default value</span>



```python
def nationality(country = 'India'):
    print('I am from',country)

nationality()
nationality('Brazil')
nationality('USA')
nationality('Australia')
nationality('China')
```

### <span style='color:orange'>Passing List as an Argument</span>


```python
def food(list):
    for x in list:
        print(x)

fruits = ['apple','banana','cherry']
food(fruits)
```

```python
def food_with_a(new_list):
     list_with_a = [x for x in new_list if 'a' in x]
     print(list_with_a)

fruits2 = ['banana','cherry','apple','beetroot','melon']
food_with_a(fruits2)
```

```python
def uppecase_elements(random_list):
    uppercase_list = [x for x in random_list if any(c.isupper() for c in x)]
    print(uppercase_list)

a_list = ['apple','Banana']
uppecase_elements(a_list)

# here c is the character of a list element.
```

### <span style='color:orange'>Positional-Only Arguments</span>

<span style='color:yellow'>You can specify that a function can have ONLY positional arguments, or ONLY keyword arguments.</span>

<span style='color:yellow'>To specify that a function can have only positional arguments, add , / after the arguments</span>

<span style='color:yellow'>argument to be passed normally, without key = value syntax</span>

```python
def greet(name,/):
    print('Hello,',name)

greet('Dhyey')
# we can't do this: 
# greet(name='dhyey')
```

```python
# normal function
def greet(name):
    print('Hello,',name)

greet('Dhyey')
```

### <span style='color:orange'>Keyword-Only Arguments</span>

<span style='color:yellow'>To specify that a function can have only keyword arguments, add *, before the arguments</span>

<span style='color:yellow'>argument to be passed normally, with key = value syntax only</span>




```python
def greet(*,name):
    print('hello,',name)

greet(name='Dhyey')
# don't do this : 
# greet('Dhyey')
```

### <span style='color:orange'>Combine Positional-Only and Keyword-Only</span>

<span style='color:yellow'>You can combine the two argument types in the same function.</span>

<span style='color:yellow'>Any argument before the / , are positional-only, and any argument after the *, are keyword-only.</span>

```python
def add_all(a,b,/,*,c,d):
    print(a+b+c+d)

add_all(1,2,c=3,d=4)
```

### <span style='color:orange'>Recursion</span>

<span style='color:yellow'>Python also accepts function recursion, which means a defined function can call itself.</span>

<span style='color:yellow'>Recursion is a common mathematical and programming concept. It means that a function calls itself. This has the benefit of meaning that you can loop through data to reach a result.</span>

```python
def factorial(n,/):
    if(n==0 or n==1):
        return 1
    else:
        return (n * factorial(n-1))

factorial(5)
```
