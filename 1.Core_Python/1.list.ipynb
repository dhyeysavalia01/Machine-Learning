### <span style="color:orange">Looping through List elements </span>


```python
LIST = [rd.randint(100,200) for _ in range (7)]
LIST
```

```python
for i in LIST:
    print(i)
```

```python
for i in range (len(LIST)):
    print(LIST[i]) # thorugh indices
```

```python
# using while loop
i = 0
while i<(len(LIST)):
    print(LIST[i])
    i+=1
```

### <span style="color:orange">List Comprehension </span>

List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

newlist = [expression for item in iterable if condition == True]


```python
l1 = [rd.randint(100,125)for _ in range(5)]
l1
```

```python
new_l1 = [x for x in l1 if (x<120)]
new_l1
```

### <span style="color:orange">List Sorting </span>

```python
# list.sort()
l1
```

```python
l1.sort()
print(l1)
```

```python
# sort in descending order
# list.sort(reverse = True)
l1
```

```python
l1.sort(reverse=True)
l1
```

```python
# case sensitive sort
fruits = ['banana','Kiwi','Avocado','apple']
fruits.sort()
fruits
```

```python
# case insensitive sort
fruits2 = ["banana", "Orange", "Kiwi", "cherry"]
fruits2.sort(key=str.lower)
fruits2
```

### <span style="color:orange">List Copying</span>

You cannot copy a list simply by typing list2 = list1, because: list2 will only be a reference to list1, and changes made in list1 will automatically also be made in list2.

You can use the built-in List method copy() to copy a list.

```python
Prev_list = [rd.randint(2,20)for _ in range(4)]
Prev_list
```

```python
next_list = Prev_list.copy()
next_list
```
