# <span style='color:orange'>lambda function</span>

<span style='color:yellow'>Syntax:</span>

<span style='color:white'>lambda arguments: expression</span>

<span style='color:green'>NOTE : can have many arguments, but only single expression</span>

```python
# printing square of a number

square = lambda x: (x**2)
print(square(5))
```

```python
# addition
add_all = lambda a,b,c,d:(a+b+c+d)
w = 5
x = 4
y = 2
z = 1
add_all(w,x,y,z)
```

```python
def fn(n):
    return lambda a:(a*n)


doubler = fn(2)
print(doubler(5))
```
