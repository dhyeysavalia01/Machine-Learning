# <span style='color:orange'>For loop</span>

```python
fruits = ['apple','banana','cherry']
for x in fruits:
    print(x)
```

```python
# with index
for i in range(len(fruits)):
    print(fruits[i])
```

```python
# looping through string
string1 = 'banana'
for x in string1:
    print(x)
```

```python
for x in range(len(string1)): # with indexes
    print(string1[x])
```

```python
# break : exits the loop for a given condition
import random as rd

list1 = [rd.randint(10,99)for _ in range (7)]
list1
```

```python
# breaking the loop when first even number occured
for x in list1:
    if(x%2==0):
        break
    else:
        print(x)
```

```python
# continue : skips the iteration for given condition.
# skipping the even numbers

for x in list1:
    if(x%2!=0):
        print(x)
    else:
        continue
```

```python
# range(start,end,increment/decrement) : prints values from 0 to n-1.
for x in range(6):
    print(x)
```

```python
for x in range(2,12,2):
    print(x)
```

```python
for x in range(2,12):
    print(x)
```

```python
# nested for loops
list_A = ['a','b','c']
list_B = ['1','2','3']

for x in list_A:
    for y in list_B:
        print(x,y)
```

# <span style='color:orange'>While loop</span>

```python
# Print i as long as i is less than 6:
i = 0
while i<6:
    print(i)
    i+=1
```

```python
# Exit the loop when i is 3:
i = 0
while i<6:
    if(i == 3):
        break
    else:
        print(i)
    i = i+1
```

```python
# skip 3
i = 0
while i<6:
    i+=1
    if(i==3):
        continue
    print(i)
```

```python
# writing table of given number n

n = int(input("n = "))

for i in range(1,11):
    print(f"{n}x{i}={n*i}")
```
