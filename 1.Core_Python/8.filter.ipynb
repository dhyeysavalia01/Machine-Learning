# <span style='color:orange'>filter() function</span>

<span style='color:yellow'>Syntax:</span>

<span style='color:white'>filter(function, iterables)</span>

<span style='color:white'>The filter() function in Python is a built-in function that filters elements from an iterable (like lists, tuples, or sets) based on a given condition. </span>

```python
# find ages >= 18

import random
ages = [random.randint(10,50) for _ in range(10)]
ages
```

```python
def above_18(i):
    if i<18:
        return False
    else:
        return True
    
adults = filter(above_18,ages)

# for i in adults:
#     print(i)

print(list(adults))
```
