# <span style='color:orange'>Random Module</span>

<span style='color:white'>The Python random module provides a set of functions for generating random numbers, selecting random elements from sequences, and shuffling collections.</span>

```python
# generates a random integer between 1 and 100
import random as rd

random_number = rd.randint(1,100)
random_number
```

```python
# Random float between -10째C and 50째C
temperature = rd.uniform(-10,50)
print(f"Temperature: {temperature:.2f}째C")
```

```python
# Selecting Random Elements from a List
# random.choice() allows you to pick a random element

colors = ['red','green','blue','purple','orange','black']
print(rd.choice(colors))
```

```python
# shuffling a list
# random.shuffle(list)

cards = ['K','Q','J','A','10']
rd.shuffle(cards)
print(cards)
```

```python
# with other iterables
# Tuples
# Direcly use random.choice() OR random.sample()

options = ('option1','option2','option3')
print(rd.choice(options))
```

```python
# rolling a dice
def roll_dice():
    return rd.randint(1,6)

print("you rolled :",roll_dice())
```

```python
# coin toss
def toss():
    return rd.choice(['heads','tails'])

print("You got :",toss())
```

```python
# random password generator
import string

def generate_password(length = 8): # given default parameter
    characters = string.hexdigits + string.punctuation
    return ''.join(rd.choice(characters) for _ in range(length))

print(generate_password(10))
```
