# <span style='color:orange'>map() function</span>

<span style='color:yellow'>Syntax:</span>

<span style='color:white'>map(function, iterables)</span>

<span style='color:white'>The map() function in Python applies a given function to each item of an iterable </span>

```python
# length on each element in a tuple.
def fn(iterable):
    return len(iterable)

x = map(fn,('banana', 'apple', 'cherry'))
print(x)
print(tuple(x))
```

```python
def total_fruits(a,b):
    return a + b

y = map(total_fruits,('apple ', 'banana ', 'cherry '),('pineapple ', 'guava ', 'orange '))
print(y)
print(list(y))
```
