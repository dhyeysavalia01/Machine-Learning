# <span style='color:orange'>Collections.counter Module</span>

<span style='color:white'>They are used to count the occurrences of elements in an iterable or to count the frequency of items in a mapping.</span>

<span style='color:green'>class collections.Counter([iterable-or-mapping])</span>

```python
from collections import Counter

values = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
# creating a counter variable
counter_C = Counter(values)
print(counter_C)
```

```python
var_1 = Counter([1, 2, 2, 3, 3, 3])
var_2 = Counter({1: 2, 2: 3, 3: 1})
var_3 = Counter("mississippi")

print(var_1)
print(var_2)
print(var_3)
```

### <span style='color:orange'>Accessing Counter Elements</span>

<span style='color:white'>We can access the count of each element using the element as the key. If an element is not in the Counter, it returns 0.</span>

```python
temp = Counter([1, 2, 2, 3, 3, 3])

print(temp[0])
print(temp[1])
print(temp[2])
print(temp[3])
print(temp[4])
```

### <span style='color:orange'>Updating Counter</span>

```python
counter_a = Counter([1,2,2])
counter_a
```

```python
counter_a.update([2,2,2,2,2,3,3,])
counter_a
```

### <span style='color:orange'>Counter Methods</span>

```python
# 1. elements() : Returns an iterator over elements repeating each as many times as its count. Elements are returned in arbitrary order.

temp_a = Counter([1, 2, 2, 3, 3, 3, 5, 6, 5, 5])
items = list(temp_a.elements())
items
```

```python
# 2. most_common() : Returns a list of the n most common elements and their counts from the most common to the least. If n is not specified, it returns all elements in the Counter.

temp_b = Counter([1, 2, 2, 3, 3, 3, 5, 6, 5, 5, 5, 5, 5, 5])
common = temp_b.most_common(1) # top most common
common 
```

```python
# 3. subtract() : Subtracts element counts from another iterable or mapping. Counts can go negative.
temp_c = Counter([1, 2, 2, 3, 3, 3, 5, 6, 5, 5, 5, 5, 5, 5])
temp_c.subtract([1,1,1,1,1,2])

temp_c
```

```python
# arithmetic operations
x = Counter([1,2,2,3])
y = Counter([2,3,3,4])

# addition
print(x+y)

# subtraction
print(x-y)

# union
print(x | y)

# intersection
print(x & y)
```
